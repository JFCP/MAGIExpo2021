```python


def ShapeAsList(w):
    if w=='nan':
        w = ''
    new = w.split(')(')
    new = [w.replace('(','').replace(')','') for w in new]
    new = [w.split(' ') for w in new]
    new = [item for sublist in new for item in sublist]
    
    new = [dict(zip('instagram', item)) if ('instagram' in item) else item for item in new]   

    # vimeo, artstation, youtube, behance, linkedin, facebook, itch.io, twitch, github
    # if it doesn't include instagram, twitter then label as 'portfolio' (including magistudio.net)
    print(new)    
    return new
    

df1['socialmedia']= df1['websites'].apply(lambda x: ShapeAsList(str(x)))
```
